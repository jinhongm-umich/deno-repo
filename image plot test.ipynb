```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

# 이미지 불러오기
img = mpimg.imread('4473_SC1_z=32mm.png')
img2 = mpimg.imread('4473_SC1_z=32mm.png')


# 이미지를 RGB로 변환 (투명도를 제거하고 배경색을 검은색으로 설정)
if img.ndim == 3 and img.shape[2] == 4:
    r, g, b, a = img[:, :, 0], img[:, :, 1], img[:, :, 2], img[:, :, 3]
    img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    img[:, :, 0] = r * a + (1 - a) * 0  # Red 채널
    img[:, :, 1] = g * a + (1 - a) * 0  # Green 채널
    img[:, :, 2] = b * a + (1 - a) * 0  # Blue 채널
elif img.ndim == 2:  # Grayscale 이미지인 경우
    img = np.stack((img, img, img), axis=-1)
elif img.ndim == 3 and img.shape[2] == 3:  # 이미 RGB 이미지인 경우
    pass
else:
    raise ValueError("Unexpected image format")

# 이미지 플롯
plt.imshow(img)
plt.axis('off')  # 축을 숨김
plt.show()
```
